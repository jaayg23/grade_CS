```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('royalties_data.csv')
```

```python
data.head()
```

```python
print("\nValores nulos:", data['Año'].isna().sum())
```

```python
data = data.dropna(subset=['Año'])
```

```python
data['Año'] = data['Año'].astype(int)
```

```python
# Opción 1: Convertir columnas específicas (excluyendo 'Año')
for column in data.columns[1:]:  # Excluye la primera columna (Año)
    data[column] = pd.to_numeric(data[column].str.replace('$', '').str.replace(',', ''), errors='coerce')
```

```python
plt.figure(figsize=(12, 6))

for column in data.columns[1:]:  
    plt.plot(data['Año'], data[column], marker='o', label=column)

plt.title('Royalties Over Years')
plt.xlabel('Year')
plt.ylabel('Royalties')

plt.legend(title='Royalties Type')
plt.grid(True)
plt.show()
```
