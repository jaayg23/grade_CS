```python
def actualizar_series_de_tiempo(diccionario, carpeta='artists_data_series'):
    if not os.path.exists(carpeta):
        os.makedirs(carpeta)
        
    # Obtener solo los IDs de la lista
    ids = list(diccionario.values())
    fecha_captura = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    
    try:
        # Petición en lote (Soporta hasta 50 IDs por llamado)
        respuesta = sp.artists(ids)
        artistas_data = respuesta['artists']
        
        for data in artistas_data:
            nombre_limpio = data['name'].replace(' ', '_').lower()
            ruta_archivo = os.path.join(carpeta, f"{nombre_limpio}.csv")
            
            # Estructurar nueva fila
            nueva_fila = pd.DataFrame([{
                'timestamp': fecha_captura,
                'popularidad': data['popularity'],
                'seguidores': data['followers']['total']
            }])
            
            # Guardar o Adjuntar
            if not os.path.isfile(ruta_archivo):
                nueva_fila.to_csv(ruta_archivo, index=False)
            else:
                nueva_fila.to_csv(ruta_archivo, mode='a', header=False, index=False)
                
        print(f"Éxito: {len(artistas_data)} artistas actualizados el {fecha_captura}")

    except Exception as e:
        print(f"Error en la extracción: {e}")

# Ejecución
actualizar_series_de_tiempo(artists_dict)
```
